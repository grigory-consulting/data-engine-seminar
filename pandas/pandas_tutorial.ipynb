```python
from faker import Faker
import random
import pandas as pd 
fake = Faker("de_DE")
```

```python
def random_gender():
    return random.choice(["männlich", "weiblich", 'divers'])

def random_blood_type():
    return random.choice(["A+", "A-", "B+", "B-","O+", "O-","AB+","AB-"])

def random_smoker():
    return random.choice([True, False])

def random_allergies():
    return random.choice(["keine", "Pollen", "Nüsse", "Milch", "Gluten"])

def random_family_status():
    return random.choice(["ledig", "verheiratet", "geschieden", "verwitwet"])
```

```python
patienten = []

for _ in range(50000):
    patient = {
    "PatientenID": fake.unique.uuid4(),
    "Versichertennummer": fake.bothify("??#######").upper(), # zwei Mal Buchstabe, 6 mal Zahl
    "Vorname": fake.first_name(),
    "Nachname": fake.last_name(),
    "Geburtsdatum": fake.date_of_birth(minimum_age=1, maximum_age=99),
    "Geschlecht": random_gender(),
    "Adresse": fake.address().replace('\n', ', '), # Bei Adresse die Neuzeilen durch Komma+Leerzeichen ersetzen
    "Telefon": fake.phone_number(),
    "Email": fake.unique.email(),
    "Blutgruppe": random_blood_type(),
    "Letzer_Besuch": fake.date_this_year(),
    "Hausarzt": f"{fake.first_name()} {fake.last_name()}",
    "Größe_cm":  random.randint(140, 220),
    "Gewicht_kg": random.randint(40, 130),
    "Raucher": random_smoker(),
    "Allergien": random_allergies(),
    "Letzes_EKG": fake.date_this_year(),
    "BMI": round(random.uniform(18,36),1), # Zufallszahl 
    "Familienstand": random_family_status(),
    "Beruf": fake.job(),
}
    patienten.append(patient)

patienten
```

```python
df = pd.DataFrame(patienten)
df
```

```python
n_df = df[["Vorname","Nachname","Geburtsdatum"]].copy() # tiefe Kopie
n_df["Alter"] = pd.Timestamp("now").year - pd.to_datetime(n_df["Geburtsdatum"]).dt.year
n_df["Geschlecht"] = df["Geschlecht"]
n_df["Raucher"] = df["Raucher"]
# Frauen zwischen 18 und 30, Nichtraucherin
n_df[(n_df["Alter"] >= 18) & (n_df["Alter"] < 30) & (n_df["Geschlecht"] == "weiblich") & (~ n_df["Raucher"])].head()
```

```python
n_df["BMI"] = df["BMI"]

def bmi_class(bmi):
    if bmi < 18.5:
        return "Untergewicht"
    elif bmi < 25:
        return "Normalgewicht"
    elif bmi < 30:
        return "Übergewicht"
    else:
        return "Adipositas"

n_df["BMI_Klasse"] = n_df["BMI"].apply(lambda bmi: bmi_class(bmi))

# Aufgabe: 
# Berechnen Sie die Altersgruppen als neue Spalte in n_df
# "Kind" (unter 18), "Erwachsen", "Senior" (ab 65)

def altersgruppe(alter):
    if alter<18:
        return "Kind"
    elif alter < 65:
        return "Erwachsen"
    else:
        return "Senior"

n_df["Altersgruppe"] = n_df["Alter"].apply(lambda alter: altersgruppe(alter))
n_df
```

```python
def filter_normalgewicht(df):
    return df[df["BMI_Klasse"] == "Normalgewicht"]

def filter_altersgruppe(df,altersgruppe):
    return df[df["Altersgruppe"] == altersgruppe]



def blutgruppe_part(df,col):
    df["Blutgruppe"] = col
    return df[(df["Blutgruppe"]=="0+") | (df["Blutgruppe"] == "AB+")]



df_normalgewicht = (n_df
    .pipe(filter_normalgewicht)
    .sort_values("Alter", ascending =False)
)
# Data Transformation Pipeline
df_normerwachsen = (n_df
    .pipe(filter_normalgewicht)
    .pipe(filter_altersgruppe, "Erwachsen")
    .sort_values("Alter", ascending =False)
)
df_normerwachsen
```
